label encoding

```python
from sklearn.preprocessing import LabelEncoder
import joblib

label_encoders = {}

# Function to encode categorical columns and store encoders
def label_encode_columns(df, categorical_columns):
    for col in categorical_columns:
        encoder = LabelEncoder()
        df[col] = encoder.fit_transform(df[col])
        label_encoders[col] = encoder
    return df

# Apply the encoding function
data = label_encode_columns(data, categorical_columns)
```

```python
data
```

```python
joblib.dump(label_encoders, 'label_encoders.pkl')
```

```python
from sklearn.preprocessing import LabelEncoder
encoder_churn = LabelEncoder()
data['Churn'] = encoder_churn.fit_transform(data['Churn'])
```

```python
joblib.dump(encoder_churn, 'encoder_churn.pkl')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data[['Monthly Charges','tenure','Total Charges']] = scaler.fit_transform(data[['Monthly Charges','tenure','Total Charges']])
```

```python
joblib.dump(scaler, 'scaler.pkl')
```

spliting the dataset

```python
X = data.drop(["Churn"],axis=1)
y = data["Churn"]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=4)
```

Models training: logitistic,decision tree and Random forest

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# Initialize models
logistic_model = LogisticRegression(penalty="l2",C=10)
decision_tree_model = DecisionTreeClassifier()
random_forest_model = RandomForestClassifier(n_estimators=100)

# Train models
logistic_model.fit(X_train, y_train)
decision_tree_model.fit(X_train, y_train)
random_forest_model.fit(X_train, y_train)

# Predict on training data
y_train_pred_logistic = logistic_model.predict(X_train)
y_train_pred_tree = decision_tree_model.predict(X_train)
y_train_pred_forest = random_forest_model.predict(X_train)

# Predict on test data
y_test_pred_logistic = logistic_model.predict(X_test)
y_test_pred_tree = decision_tree_model.predict(X_test)
y_test_pred_forest = random_forest_model.predict(X_test)
```

```python
# Function to print evaluation metrics
def print_metrics(y_true, y_pred, dataset_type):
    print(f"Metrics for {dataset_type} dataset:")
    print(f"Accuracy: {accuracy_score(y_true, y_pred):.4f}")
    print(f"Precision: {precision_score(y_true, y_pred):.4f}")
    print(f"Recall: {recall_score(y_true, y_pred):.4f}")
    print(f"F1 Score: {f1_score(y_true, y_pred):.4f}")
    print(confusion_matrix(y_true, y_pred))
    print("-" * 30)

# Evaluate models on training data
print("Logistic Regression")
print_metrics(y_train, y_train_pred_logistic, "Training")
print_metrics(y_test, y_test_pred_logistic, "Test")

print("Decision Tree")
print_metrics(y_train, y_train_pred_tree, "Training")
print_metrics(y_test, y_test_pred_tree, "Test")

print("Random Forest")
print_metrics(y_train, y_train_pred_forest, "Training")
print_metrics(y_test, y_test_pred_forest, "Test")
```

the logistics regression performs best in this situation because the difference between the test and training is negligeable:
Accuracy training: 0.8048
Accuracy test: 0.7950

the random forest has the second best performance with:
Accuracy training: 0.9984
Accuracy tets: 0.7850

for the decision tree we notice an overfitting with:
Accuracy training: 0.9984
Accuracy test: 0.7347



the best model to deploy is the logistics regression because the training and test accuracy are close to each other
